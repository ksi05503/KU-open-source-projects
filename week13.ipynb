```python
import csv
import matplotlib.pyplot as plt

# 2011.01.01 ~ 2020.12.31 서울 기온 데이터
f = open('seoul1.csv', 'r', encoding='cp949')
data = csv.reader(f)

next(data)
# 날짜, 지점, 평균기온,  최저기온, 최고기온
#  0     1       2          3         4

high = [] # 최고 기온 누적 리스트
low = [] # 최저 기온 누적 리스트

for row in data:
    if row[-1] != '' : # 빈 데이터 스킵
        if (row[0].split('-')[1] == '10') and (row[0].split('-')[2] == '10'):
            high.append(float(row[4])) # 최고기온 high에 append
            low.append(float(row[3])) # 최저기온 low에 append
            
plt.plot(high, color='pink')#, label='HIGH')
plt.plot(low, color='skyblue')#, label='LOW')
plt.show() 
```

```python

import csv
import matplotlib.pyplot as plt

f = open('seoul1.csv', 'r', encoding='cp949')
data = csv.reader(f, delimiter=',')
header = next(data)
janu = []
augu = []
for row in data:
    if row[0].split('-')[1] == '01':
        janu.append(float(row[-1]))
    if row[0].split('-')[1] == '08':
        augu.append(float(row[-1]))
plt.rc('font', family='Malgun Gothic')
plt.title('서울 지역 1월과 8월의 최고 기온 히스토그램')
plt.hist(janu, bins=100, color='blue', label='1월')
plt.hist(augu, bins=100, color='red', label='8월')
plt.legend()
plt.show()
plt.close()
```
